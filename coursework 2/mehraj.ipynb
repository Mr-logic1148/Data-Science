```python
                                        """RETRIVING DATA FROM THE DATASET"""
    
dataset = pd.read_csv('COVID_Symptom_Checker.csv')
print(dataset.info())
#null_values = dataset.isnull().sum()
#print(null_values)
signs_and_symptoms = dataset.iloc[:, :22].values
symptoms = dataset.iloc[:, 0:6].values
signs = dataset.iloc[:, 6:10].values
age_gender = dataset.iloc[:, 11:19].values
contact = dataset.iloc[:, 19:22].values
severity = dataset.iloc[:, 22:].values
#print(signs_and_symptoms)
print(symptoms)
#print(age_gender)
#print(contact)
```

```python
                                            """SEVERITY LABEL"""
le = LabelEncoder()
severity_label = le.fit_transform(severity)
print("""SEVERITY VALUES:""")
print(severity)
print("SEVERITY LABELS:")
print(severity_label)
print("*******************")
```

```python
                                    """TO CHECK PREDICTION FOR SYMPTOMS:"""

#FEVER COLUMN
fever_symptom = symptoms[:, 0:1]
fever_symptom = fever_symptom.reshape(316800)
print("\n FEVER COLUMN:")
#print(fever.shape)
print(fever_symptom)
fever_symptom = le.fit_transform(fever_symptom)
print("FEVER LABELS:")
print(fever_symptom)
print("*******************")

#TIREDNESS
tiredness_symptom = symptoms[:, 1:2]
tiredness_symptom = tiredness_symptom.reshape(316800)
print("\n TIREDNESS COLUMN:")
#print(fever.shape)
print(tiredness_symptom)
tiredness_symptom = le.fit_transform(tiredness_symptom)
print("TIREDNESS LABELS:")
print(tiredness_symptom)
print("*******************")

#DRY COUGH COLUMN
dry_cough_symptom = symptoms[:, 2:3]
dry_cough_symptom = dry_cough_symptom.reshape(316800)
print("\n DRY COUGH COLUMN:")
#print(DRY_COUGH.shape)
print(dry_cough_symptom)
dry_cough_symptom = le.fit_transform(dry_cough_symptom)
print("DRY COUGH LABELS:")
print(dry_cough_symptom)
print("*******************")

#DIFFICULTY IN BREATHING 
difficulty_in_breathing_symptom = symptoms[:, 3:4]
difficulty_in_breathing_symptom = difficulty_in_breathing_symptom.reshape(316800)
print("\n DIFFICULTY IN BREATHING COLUMN:")
#print(fever.shape)
print(difficulty_in_breathing_symptom)
difficulty_in_breathing_symptom = le.fit_transform(difficulty_in_breathing_symptom)
print("DIFFICULTY IN BREATHING LABELS:")
print(difficulty_in_breathing_symptom)
print("*******************")

#SORE THROAT
sore_throat_symptom = symptoms[:, 4:5]
sore_throat_symptom = sore_throat_symptom.reshape(316800)
print("\n SORE THROAT COLUMN:")
#print(fever.shape)
print(sore_throat_symptom)
sore_throat_symptom = le.fit_transform(sore_throat_symptom)
print("SORE THROAT LABELS:")
print(sore_throat_symptom)
print("*******************")

#NONE SYMPTOM
none_symptoms = symptoms[:, 5:6]
none_symptoms = none_symptoms.reshape(316800)
print("\n NONE SYMPTOM COLUMN:")
#print(fever.shape)
print(none_symptoms)
none_symptoms = le.fit_transform(none_symptoms)
print("NONE SYMPTOM LABELS:")
print(none_symptoms)
print("*******************")

#PREDICTION ARRAY FOR SYMPOTMS COLUMN IN THE DATASET
predictive_symptoms_array = np.array([fever_symptom, tiredness_symptom, dry_cough_symptom, difficulty_in_breathing_symptom, sore_throat_symptom,none_symptoms])
predictive_symptoms_array = predictive_symptoms_array.reshape(316800, 6)
print(predictive_symptoms_array.shape)
print(predictive_symptoms_array)
```

```python
                                         """PREDICTION TRAIN-TEST FOR SYMPTOMS"""

symptom_train, symptom_test, sy_train, sy_test = train_test_split(predictive_symptoms_array, severity_label, test_size = 0.7, random_state = 43)

print(symptom_train)
print("*****************")
print(symptom_test)
```

```python
                                            """PREDICTION FOR SYMPTOMS"""
    
symptom_clf = CategoricalNB()
symptom_clf.fit(symptom_train, sy_train)

symptom_prediction = symptom_clf.predict(symptom_test)
print(symptom_prediction)
    
```

```python
                                         """ACCURACY FOR SYMPTOMS"""
    
symptom_accuracy = accuracy_score(sy_test, symptom_prediction)
print(symptom_accuracy)
```

```python
                                    """ERROR RATE FOR SYMPTOMS"""
    
symptom_error_rate = 1 - symptom_accuracy
print(symptom_error_rate)
```
